```python
import numpy as np
import cv2
from skimage.measure import block_reduce
import matplotlib.pyplot as plt

resim=cv2.imread("/content/images.jpeg")
resim=cv2.cvtColor(resim, cv2.COLOR_BGR2GRAY)

filtre=np.array([[0,-1,0], [-1,5,-1], [0,-1,0]])
filtre_resim=cv2.filter2D(resim, -1, filtre)
relu=np.maximum(filtre_resim, 0)

pool_size=(2,2)
pool_resim=block_reduce(relu, pool_size, np.max)


plt.figure(figsize=(15, 5))

plt.subplot(1,3,1)
plt.title("Original Image")
plt.imshow(resim,  cmap="gray")

plt.subplot(1,3,2)
plt.title("ReLU Image")
plt.imshow(relu, cmap="gray")

plt.subplot(1,3,3)
plt.title("Pooling Image")
plt.imshow(pool_resim, cmap="gray")

plt.tight_layout()
plt.show()


```
