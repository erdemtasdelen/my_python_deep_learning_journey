```python
from keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions

model=VGG16()
model.summary()
```

```python
from tensorflow.keras.preprocessing import image
resim=image.load_img("/content/Unknown.jpeg", color_mode="rgb", target_size=(224,224))
display(resim)
```

```python
import numpy as np

resim=image.img_to_array(resim)
resim=np.expand_dims(resim, axis=0)


prediction=model.predict(resim)
prediction
```

```python
p=decode_predictions(prediction)
p
```

```python
import matplotlib.pyplot as plt

resim=preprocess_input(resim)

label=[i[1] for i in p[0]]
probability=[i[2] for i in p[0]]
plt.bar(label, probability)
plt.title("Prediction")
plt.xticks(rotation=90)
plt.show()
```
